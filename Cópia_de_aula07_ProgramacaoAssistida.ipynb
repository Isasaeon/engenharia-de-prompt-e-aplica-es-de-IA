Soma de Lista

```python
# Define uma lista de números inteiros
numeros = [1, 3, 6, 7, 8]

# Inicializa uma variável para armazenar a soma dos números
soma = 0

# Itera sobre cada número na lista 'numeros'
for num in numeros:
  # Adiciona o número atual à variável 'soma'
  soma += num

# Imprime o valor final da soma de todos os números
print(soma)
```

Eh primo

```python
import math

# Define a function to check if a number is prime
def eh_primo(numero):
    """
    Verifica se um número é primo.
    Um número primo é um número natural maior que 1 que não pode ser formado pela multiplicação de dois números naturais menores.
    """
    if numero <= 1:
        return False  # Números menores ou iguais a 1 não são primos
    if numero == 2:
        return True   # 2 é o único número primo par
    if numero % 2 == 0:
        return False  # Qualquer outro número par não é primo

    # Verifica divisibilidade por números ímpares a partir de 3 até a raiz quadrada do número
    # Não precisamos verificar divisores maiores que a raiz quadrada, pois se houvesse um,
    # haveria também um menor que a raiz quadrada.
    for i in range(3, int(math.sqrt(numero)) + 1, 2):
        if numero % i == 0:
            return False  # Se for divisível, não é primo
    return True       # Se não encontrou divisores, é primo

# Início do bloco try para capturar erros de entrada
try:
    # Captura o valor do usuário e converte para inteiro
    numero = int(input("Digite um número inteiro para verificar se é primo: "))

    # Chama a função eh_primo para verificar o número
    if eh_primo(numero):
        # Exibe mensagem se o número for primo
        print(f"O número {numero} é primo! ✨")
    else:
        # Exibe mensagem se o número não for primo
        print(f"O número {numero} não é primo. ❌")
```

Inversão String

```python
def inverte_texto(texto):
  # Esta função recebe uma string e a retorna invertida
  # Utilizamos o fatiamento (slicing) [::-1] que é a forma mais eficiente em Python
  return texto[::-1]

# Exemplo de uso:
nome = "isadora"
resultado = inverte_texto(nome)

# Exibe o resultado da inversão
print(f"Original: {nome}")
print(f"Invertido: {resultado}")
```
